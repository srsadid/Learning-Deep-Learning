```python
import pydicom as dicom
import cv2   
import os 

# specify your image path

target_dir = r'F:\pneumonia\000 P R O J E C T S\pneumonia\New folder\data'
img_dir = r'F:\pneumonia\000 P R O J E C T S\pneumonia\New folder'

filename = 'sample.dcm'

image_path = os.path.join( img_dir , filename)
target_path = os.path.join( target_dir , filename)
ds = dicom.dcmread(image_path)

pixel_array_numpy = ds.pixel_array

image_format = '.jpg' # or '.png'
image = target_path.replace('.dcm', image_format)

cv2.imwrite( image , pixel_array_numpy)
```
